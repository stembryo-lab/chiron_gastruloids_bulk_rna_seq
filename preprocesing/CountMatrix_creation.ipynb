#### Alignment files are concatenated into a matrix

```python
import os
import numpy as np
from numpy import isneginf
import pandas as pd
import glob
import sys
import csv
import re
```

```python
path = 'STAR_Results/Gene_counts/' 

files = glob.glob(os.path.join(path , "*.tab"))
```

```python
filenames=[]
for filename in files:
    filename=filename.split('/')[2].split('.tab')[0]
    filenames.append(filename)
```

```python
filenames
```

```python
metadata=pd.read_csv('Data/metadata_andre.csv')
metadata[:5]
```

```python
order=metadata['ID'].to_list()
order
```

```python
# set order using metadata
order=metadata['ID'].to_list()

sorted_files=sorted(filenames, key=lambda x: order.index(x))
sorted_files[:5]
```

```python
li = []
samples =[]

#files=sorted(files, key=multiple_sort)

for filename in sorted_files:
    df = pd.read_csv('STAR_Results/Gene_counts/'+filename+'.tab', index_col=0, header=0, sep='\t', skiprows=3)
    df = df.iloc[:,1]

    li.append(df)

counts = pd.concat(li, axis=1, ignore_index=True)
counts = counts.loc[~(counts==0).all(axis=1)]
counts.columns = metadata['Sample']
counts.index.name = 'Gene_id'
counts = counts.fillna(0)


# save df
counts.to_csv(path + 'counts_chir_af_project.csv')
counts
```
